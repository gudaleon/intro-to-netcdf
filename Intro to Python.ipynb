# Intro to Python

As a demo of the language, plot sin and cos on the same axis:

```python
# load some key packages:
import numpy as np                # a numerical package
import matplotlib.pyplot as plt   # a plotting package

# to display the plots in the same document
%matplotlib inline
```

```python
# define an array of points (start, end, by)
x = np.linspace(0,6*np.pi,100)

# plot the sin of these points
plt.plot(np.sin(x), 'b')

# plot the cos of these points
plt.plot(np.cos(x),'r')
```

# Netcdf

The simplest way to access netcdf files using python is by installing the [xarray package](http://xarray.pydata.org/en/stable/, "xarray docs"). There are other packages that can be used, but for now I am going to concentrate on xarray

```python
import xarray as xr

data_url = 'http://hydromet-thredds.princeton.edu/thredds/dodsC/MonitoringStations/CR1000_Table1.nc'
variable = 'AirTC_Avg'

ds = xr.open_dataset(data_url)
```

```python
ds[variable].plot()
```
